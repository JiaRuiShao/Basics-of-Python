Download stock data for Apple for the last three months into a csv file (from Yahoo finance). Open the csv in excel and remove all columns except closing price. Write python code to import the closing prices into an ndarray. Calculate the % daily change in price.


```python
import numpy as np
import csv

open_file=open("APPLE.csv", "r")

myList=list(open_file)
arr=np.array(myList).astype(np.float64)
print(arr,'\n')

arr1=arr.copy()
arr1=arr1[1:]
print(arr1,end='\n')

arr2=arr.copy()
arr2=arr2[:-1]
print(arr2)

change=arr1-arr2
np.set_printoptions(precision=6)
print(change)

change_percentage=change*100/arr2
print(change_percentage)
```
