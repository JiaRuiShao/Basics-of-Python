```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Load data Video Game Sales**

Data source: Kaggle [https://www.kaggle.com/gregorut/videogamesales/download]

Big idea: The video game industry reached its peak in 2010 and starts to decrease in sales after that, with almost all games' global sales are from 0 to 20.


```python
data = pd.read_csv('vgsales.csv')
data.head()
```

**Line Chart**

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
data.groupby(['Year']).count()['Global_Sales'].plot(ax=ax)
```

**Pie Chart**

```python
platform = data['Platform'].value_counts()

platform.plot.pie(figsize=(15, 15))
```

**Bar Chart**

```python
genre = data['Genre'].value_counts()
genre.plot.bar(figsize=(15,10))
```

**Histogram**

```python
plt.hist(data['Global_Sales'], bins=30)
plt.show()
```
