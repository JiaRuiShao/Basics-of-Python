Here is the link to official student evaluations of CIS professors and courses from Spring 2017. Scrape this page and get a list of courses, section numbers, and names of professors for which we have the evaluations posted.

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests

r = requests.get('http://www.baruch.cuny.edu/ote/evals/evaluations/2017SPRING/CIS.html')
c=r.text
soup = BeautifulSoup(c, 'html.parser')
#print(soup)
```

```python
x = soup.findAll('a')
for anchor in x:
    if anchor.get('href') != None:
        if '.pdf' in anchor.get('href'):
            print(anchor.get_text())           
```

Here is the link to the faculty directory for CIS. Scrape this page and list faculty names, email addresses, and their web URLs.

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests

r = requests.get('https://cis.baruch.cuny.edu/faculty/')
c=r.text
soup = BeautifulSoup(c, 'html.parser')
#print(soup)
```

```python
x = soup.findAll('a')
x=x[3:]
print("The professors who have the link to their profile:",end='\n\n')
for anchor in x:
    if anchor.get('href') != None:
        if 'http://' in anchor.get('href'):
            print(anchor.get_text().strip(),anchor.get('href'))  
print("\n")
print("All the professors' email':",end='\n\n')
for anchor in x:
    if anchor.get('href') != None:
        if '@' in anchor.get('href'):
            print(anchor.get_text().strip())
```
