The decimal number, 585 = 10010010012 (binary), is palindromic in both bases. Find all numbers under 1000 that are palindromic in base 10 and base 2. (Please note that the palindromic number, in either base, may not include leading zeros.) From https://projecteuler.net/problem=36

```python
for n in range (1,1001):
    string_base_2=''
    string_base_10=''
    i=n
    j=n
    
    while i != 0:        
        string_base_10 += str(i%10)
        i = i//10
    
    if string_base_10[::-1] == string_base_10:
        while j != 0:
            string_base_2 += str(j%2)
            j = j//2
        if string_base_2[::-1] == string_base_2:
            print(n,'is a palindromic number in base 10 and base 2.')
```

A palindromic prime is a number that is simultaneously palindromic and prime. Find palindromic prime numbers under 1000. From http://mathworld.wolfram.com/PalindromicPrime.html

```python
#What is the base of the palindromic number?
#decimal???
#The first few decimal palindromic primes are:
#2, 3, 5, 7, 11, 101, 131, 151, 181, 191, 313, 353, 373, 383, 727, 757, 787, 797, 919, 929, …
for i in range (1,1000):
    if str(i) == str(i)[::-1]:
        #then i is a  palindromic number.
        #find whether it's also a prime number or not
        is_prime=True
        if i < 2:
            is_prime=False
        else:
            for j in range (2,i):
                if i%j == 0:
                    is_prime=False
            if is_prime:
                print(i,'is a palindromic prime number')
```

145 is a curious number, as 1! + 4! + 5! = 1 + 24 + 120 = 145. That is, the sum of the factorial of each of its digits equals the number itself. Write code to display all numbers below 1000 for which this is true. (30 points)

```python
#There only have one curious number under 1000?
#Is there any better way to solve this problem?
def sum_of_factorial(n):
    total=1
    for x in range (1,n+1):
        total*=x
    return total

def find_curious_number():
    i=1
    j=1
    h=1
    while(sum_of_factorial(i)+sum_of_factorial(j)+sum_of_factorial(h) < 1000):
        for i in range (1,1000):
            for j in range(1,1000):
                for h in range(1,1000):
                    summ=sum_of_factorial(i)+sum_of_factorial(j)+sum_of_factorial(h)
                    if str(summ) == str(i)+str(j)+str(h):
                        print(str(summ)+'='+str(i)+'!'+str(j)+'!'+str(h)+'!')
    return #str(summ)+'='+str(i)+'!'+str(j)+'!'+str(h)+'!'

find_curious_number()
```

```python
#Better solution
def sum_of_factorial(n):
    total=1
    if n == 0:
        return 1
    if n > 0:
        for x in range (1,n+1):
            total*=x
        return total

def main():
    for i in range(1000):
        mystring=str(i)
        mylength=len(mystring)
        total=0
        for j in range (mylength):
            mynumber=int(mystring[j])
            total += sum_of_factorial(mynumber)
        if mystring == str(total):
            print(i)                                      

main()
```
