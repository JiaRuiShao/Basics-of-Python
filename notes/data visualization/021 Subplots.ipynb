```python
# plots reside within a figure.  Start by creating a figure.
fig = plt.figure()

# can't make a plot with an empty figure.  Add subplots
# Up to four plots in total.  
# Select the first of four subplots.
ax1 = fig.add_subplot(2, 2, 1)

ax2 = fig.add_subplot(2, 2, 3)

ax3 = fig.add_subplot(2, 2, 4)


# draw on the last figure and subplot used
plt.plot(randn(50).cumsum(), 'k--')
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

plt.plot(randn(50).cumsum(), 'k--') #black dashed line
_ = ax1.hist(randn(100), bins=20, color='k', alpha=.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
```

```python
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

#for i in range(2):
#    for j in range(2):
#        axes[i, j].hist(randn(500), bins=50, color='k', alpha=.5)
        
axes[0, 0].hist(randn(500), bins=50, color='k', alpha=.5)
axes[0, 1].hist(randn(500), bins=50, color='k', alpha=.5)
axes[1, 0].hist(randn(500), bins=50, color='k', alpha=.5)
axes[1, 1].hist(randn(500), bins=50, color='k', alpha=.5)

plt.subplots_adjust(wspace=0, hspace=0)
```

### Bar Plots from Series
Vertical bar plots are created with ***plot.bar()***  
  
Horizontal bar plots are created with ***plot.barh()***

```python
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

fig, axes = plt.subplots(2, 1)

# index is x-axis ticks
data.plot.bar(ax=axes[0], alpha=.5)

# index is y-axis ticks
data.plot.barh(ax=axes[1], color='k', alpha=0.5)
```
