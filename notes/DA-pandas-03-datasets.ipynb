```python
#concatenating dataframes
#https://pandas.pydata.org/pandas-docs/stable/merging.html
#In this example you have Campaign finance data split into two files. How do you bring them together into one dataframe?

d1=pandas.read_csv('Campaign-1.csv', low_memory=False)
d2=pandas.read_csv('Campaign-2.csv', low_memory=False)

d3=pandas.concat([d1,d2])
d3[23:35]



```

```python
#Another way to do the same thing (but with default axis)
d4=d1.append(d2)
d4.info()
```

```python
#What happens in the data is split into two files with one overlapping key column? The two files have different columns 
#that can be connected using the overlapping key column (in this case the ID column in both the files). 
#more on concat

d11=pandas.read_csv('Campaign-Split-1.csv', low_memory=False)
d12=pandas.read_csv('Campaign-Split-2.csv', low_memory=False)

result =pandas.concat([d11,d12], axis=1, join='inner')

result.head()
```

```python
#
result = pandas.merge(d11, d12, on='ID')

result.info()
```
