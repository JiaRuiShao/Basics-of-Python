### Data Process

```python
df = pd.read_csv('examples/HAM10000_metadata.csv')
df.head(10)
```

```python
df.isnull().sum()
```

### Explanatory Analysis

```python
df.groupby('dx').size().plot.barh()
```

```python
df['dx'].value_counts().plot(kind='barh')
```

```python
sns.distplot(df.age.dropna()).set_title("Age Distribution")
```

```python
sns.catplot(x="dx", y="age", col="sex", kind="box", data=df)
```

```python
# Distribution of males and females
df['sex'].value_counts().plot(kind='bar')
```

```python
sns.countplot(y="dx", hue="sex", data=df)
```

```python
sns.catplot(x="dx", hue="sex", col="dx_type", data = df, kind="count", height=4, aspect=1.0)
```
