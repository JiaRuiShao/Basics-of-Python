```python
import pandas
import numpy
import matplotlib.pyplot as plt
import seaborn
```

```python
CI_url="https://data.cityofchicago.org/resource/6zsd-86xi.csv"
data=pandas.read_csv(CI_url, low_memory=False)

NY_url="https://data.cityofnewyork.us/resource/9s4h-37hy.csv" 
data1=pandas.read_csv(NY_url, low_memory=False)
```

```python
dfb=pandas.DataFrame(data, columns=['year', 'primary_type'])                       #Chicago
dfb1=pandas.DataFrame(data1, columns=['cmplnt_fr_dt', 'ofns_desc'])     #NY
```

```python
rad_count = dfb['primary_type'].value_counts()           #THIS IS CHICAGO
print(rad_count)
```

```python
rad_count1 = dfb1['ofns_desc'].value_counts()              #THIS IS NEW YORK
print(rad_count1)
```

```python
%matplotlib inline
g=seaborn.countplot(x='primary_type', data=dfb,order=dfb['primary_type'].value_counts().iloc[:10].index)
g.set_xticklabels(g.get_xticklabels(),rotation=90)
```

```python
%matplotlib inline
g=seaborn.countplot(x='ofns_desc', data=dfb1,order=dfb1['ofns_desc'].value_counts().iloc[:10].index)
g.set_xticklabels(g.get_xticklabels(),rotation=90)
```
